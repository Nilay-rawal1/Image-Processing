```python
import random
import cv2
from matplotlib import pyplot as plt
```

# Function Definition 
Funtion add_noise() for adding Black and white 

```python
def add_noise(img):
    row,col=img.shape
    number_of_pixels= random.randint(300,10000)
    for i in range(number_of_pixels):
        ycord=random.randint(0,row-1)
        xcord=random.randint(0,col-1)
        img[ycord][xcord]=255
    number_of_pixels=random.randint(300,10000)
    for i in range(number_of_pixels):
        ycord=random.randint(0,row-1)
        xcord=random.randint(0,col-1)
        img[ycord][xcord]=0

    return img
#
    
```

# code to read the image

# salt and pepper effect


```python
img=cv2.imread('apple.jpeg',cv2.IMREAD_GRAYSCALE)
cv2. imwrite('applewithnoise.jpg',add_noise(img))
noise_image=cv2.imread("applewithnoise.jpg")
plt.imshow(noise_image)
plt.show()
```
