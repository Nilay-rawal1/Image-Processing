```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# reading and preprocessing


```python
img=cv2.imread('sudoku.png')
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img)
```

# hough transformation operation


```python
edges=cv2.Canny(gray,50,150,apretureSize=3)
lines=cv2.HoughLines(edges,1,np.pi/100,200)
for line in lines:
    rho,theta= line[0]
    a=np.cos(theta)
    b=np.sin(theta)
    x0=b*rho
    y0=b*rho
    x1= int(x0 +1000*(-b))
    y1= int(y0 + 1000(a))
    x2= int(x0 - 1000*(-b))
    y2= int(y0 - 1000(a))
    cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)
plt.imshow(img)
    
```
