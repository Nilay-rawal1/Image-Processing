# Importing Libraries


```python
import numpy as np
import matplotlib.pyplot as plt
```

# Reading the image


```python
import cv2
image=cv2.imread("baby.jpg")
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.title("Original")
plt.imshow(image_rgb)
```

# grayscale image


```python
gray_image=cv2.imread("baby.jpg", cv2.COLOR_BGR2GRAY)

#saving using imwrite()
cv2.imwrite("baby_gray.jpg",gray_image)
plt.title("Gray-scale")
plt.imshow(gray_image)
```

# Loading and displaying the saved image


```python
#loading section
img=cv2.imread("baby_gray.jpg")
#displaying image
plt.title("saved image")
plt.imshow(img)
```
