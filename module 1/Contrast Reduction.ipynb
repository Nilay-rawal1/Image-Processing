```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
image = cv2.imread('sample.jpg')
plt.imshow(image)
```

```python
images = cv2.bitwise_not(image)
plt.imshow(images)
cv2.imwrite('output.jpeg',images)
```
