```python
import cv2
from matplotlib import pyplot as plt
```

# open cv

```python
img= cv2.imread('opencv.png')
name='OpenCV Python Image' 
plt.title(name)
plt.imshow(img)
```

# Flipping the image horizontally

```python
horizontal_flip=cv2.flip(img,1)
name='OpenCV Flip image Horizontally'
plt.title(name)
plt.imshow(horizontal_flip)
```

#  Flipping image vertically

```python
vertical_flip=cv2.flip(img,0)
name='OpenCV Python Flip Image Vertically'
plt.title(name)
plt.imshow(vertical_flip)
```

# Flipping the image both ways


```python
verti_hori_flip=cv2.flip(img,-1)
name='OpenCV Image Flip Both Ways'
plt.title(name)
plt.imshow(verti_hori_flip)
```
